### 1) Find the average cost of each variety of product available in the three different shops
    
    Values = [[10, 20, 30], [17, 14, 21], [25, 15, 12]],
    columns = ['product_a', 'product_b', 'product_c'],
    index = ['Shop1', 'Shop2', 'Shop3']
    Convert the above into a dataframe
    
    Expected Output format: (please refere the provided image for this)

```python
import pandas as pd
Values = [[10, 20, 30], [17, 14, 21], [25, 15, 12]]
columns = ['product_a', 'product_b', 'product_c']
index = ['Shop1', 'Shop2', 'Shop3']
data = pd.DataFrame(Values, columns=columns, index=index)
data = data.T 
data.loc['Avg'] = data.mean()  

print(data.round(2))
```

### 2) Write a function to calculate the total sales for each product
    Things to keep in mind:
        • The input for the function should be the given dataframe
        • The input represents the sales records. The columns are 'Date', 'Product', 'Quantity', and 'Price'.
        • The output format is expected to be a dataframe with the product names and their total sales.
        
    Below cell is the data, you can run the cell and get the required dataframe

```python
data = {'Date': ['2023-09-01', '2023-09-01', '2023-09-02', '2023-09-02'],
        'Product': ['Laptop', 'Mouse', 'Keyboard', 'Laptop'],
        'Quantity': [2, 5, 3, 1],
        'Price': [1000, 20, 50, 100]}`
        
sales_data = pd.DataFrame(data)
```

```python
sales_data
```

```python
def total_sales(data):
    data['Sales'] = data['Quantity'] * data['Price']
    res = data.groupby('Product')['Sales'].sum()
    res.columns = ['Product', 'Total Sales']
    return res
```

```python
print(total_sales(sales_data))
```

### 4) Create a function that prints a pattern of 1's and 0's based on the given requirements
    Requirements:
        • The function should only take one user given input, which is the size of the matrix
        • When the size is less than 3, give the option to ask for input again. If the given input is wrong then
        print the text 'Size can be greater than 3 only, try again later'
        • When the given user input is odd, then print the pattern X using 1's in the pattern of 0's and 1's
        • When the given user input is even, then print the pattern Z using 0's in the pattern of 0's and 1's
        • The output doesn't strictly have to be in the given format (Can be an array, list of lists, etc)

        An example for size 3:
        1 0 1
        0 1 0
        1 0 1
        
    Please mention if any other assumptions were required/made

```python
def pattren(size):
    if size<3:
        size = int(input("enter size greater than 3..... "))
        if size < 3:
            print("size can be greater than 3 only, try again later..")
            return
            
    if size % 2 != 0:
        for i in range(size):
            for j in range(size):
                if i == j or i + j == size - 1:
                    print(1, end=" ")
                else:
                    print(0, end=" ")
            print()
    else:
        for i in range(size):
            for j in range(size):
                if i == 0 or i == size - 1 or i + j == size - 1:
                    print(0, end=" ")
                else:
                    print(1, end=" ")
            print()
```

```python
def ID_Assigning(df):
    person_id = []
    next_id = 0
    for i, r in df.iterrows():
        matched_id = None
        for j, prev_r in df.iterrows():
            if j >= i:
                break
            if r['name'] == prev_r['name'] and (
                r['father'] == prev_r['father'] or
                r['nick_name'] == prev_r['nick_name'] or
                r['birth_year'] == prev_r['birth_year']
            ):
                if matched_id is None or person_id[j] < matched_id:
                    matched_id = person_id[j]
        if matched_id is None:
            matched_id = next_id
            next_id += 1
        person_id.append(matched_id)

    res = pd.DataFrame({'Person ID': person_id})
    return res
```
